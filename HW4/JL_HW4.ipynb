```python
# Manual Tag 
#"Fierce fighting is taking place in the eastern city where Ukrainian officials said Russian artillery 
#had driven its forces out of the centre."
manual_tag = [
    ('Fierce', 'JJ'),
    ('fighting', 'NN'),
    ('is', 'VBG'),
    ('taking', 'VBG'),
    ('place', 'NN'),
    ('in', 'IN'),
    ('the','DT'),
    ('eastern', 'JJR'),
    ('city', 'NN'),
    ('where', 'WRB'),
    ('Ukranian', 'NNP'),
    ('officials','NNS' ),
    ('said','VB' ),
    ('Russian','NNP' ),
    ('artillery', 'NN'),
    ('had', 'VBD'),
    ('driven', 'VB'),
    ('its', 'PRP$'),
    ('forces', 'NNS'),
    ('out','IN'),
    ('of', 'IN'),
    ('the', 'DT'),
    ('centre', 'NN'),
    ('.', '.')
]
```

```python
# spacy's turn
news_doc = nlp(news)
spacy_tag = [(token.text, token.tag_) for token in news_doc]
print(spacy_tag)
```

```python

print(len(manual_tag))
print(len(spacy_tag))
```

```python
diffs = []
for i in range(len(spacy_tag)):
    if manual_tag[i] != spacy_tag[i]:
        diffs.append(manual_tag[i])
        diffs.append(spacy_tag[i])
    print(diffs)
```
